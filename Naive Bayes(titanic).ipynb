```python
import pandas as pd
```

```python
dataset = pd.read_csv('titanic_train.csv')
```

```python
#dat2 = pd.read_csv('titanic_test.csv')
```

```python
dataset
```

```python
# removing irrelevant features
dataset.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
dataset.head()
```

```python
inputs = dataset.drop('Survived',axis='columns')
target = dataset.Survived.fillna(dataset.Survived.mode())
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
# now lets check for blank values
inputs.columns[inputs.isna().any()]
```

```python
# We see that Age and fare contain blank values
# lets fill them with mean values
```

```python
inputs['Age'] = inputs.fillna(inputs.Age.mean())
```

```python
#inputs['Fare'] = inputs.fillna(inputs.Fare.mean())
```

```python
inputs.columns[inputs.isna().any()] # no blank values now
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import KFold
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(model,inputs,target, cv = 3) # CV gets us number of folds
```
