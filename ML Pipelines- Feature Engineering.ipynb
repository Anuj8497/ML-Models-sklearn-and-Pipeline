```python
# first we apply feature engineering on train data
# then we apply feature engineering on test data
```

```python
missing_value_feature = []
# lets see number of blank values in different categorical features
for feature in dataset.columns:
    if(dataset[feature].dtypes == 'O' and dataset[feature].isnull().sum()>0):
        print(feature)
        missing_value_feature.append(feature)
        
```

```python
data = dataset.copy()
```

```python
# whenever we have missing values in the categorical feature always create new label say 'missing for it'
for feature in missing_value_feature:
    dataset[feature].fillna('Missing',inplace = True)
```

```python
for features in missing_value_feature:
        print(features)
        print(dataset[features].isnull().sum())
```

```python
# Now lets do the same for numerical variables
numerical = [feature for feature in dataset.columns if dataset[feature].dtypes!='O']
```

```python
# checking number of blank values in numerical list
for feature in numerical:
    if(dataset[feature].isnull().sum()>0):
        print(feature)
        print(dataset[feature].isnull().sum())
    
```

```python
# From our exploratory data analysis phase, we know that there are a lot of outliers in numerical variables.
# Hence in such cases, the missing value must be replaced by median
numerical_missing_values = ['LotFrontage','MasVnrArea','GarageYrBlt']
for feature in numerical_missing_values:
    dataset[feature].fillna(dataset[feature].median(),inplace = True)
```

```python
temporal = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']
```

```python
for feature in temporal:
    if(feature != 'YrSold'):
        dataset[feature]=dataset['YrSold']-dataset[feature]
    
```

```python
dataset['YrSold']
```

```python
# from exploratory phase, we know that numerical variables are skwed. Applying log transformations to make them normal.
# we will do this on continuous variables
continuous = ['LotFrontage','LotArea' ,'1stFlrSF' ,'GrLivArea']
```

```python
for feature in continuous:
    dataset[feature]=np.log(dataset[feature])
```

```python
# Handling rare categorical features
# Say I have a categorical variable say X with levels x1,x2,x3. And if number of observations for level say x3 
# is less than 1 percent of observations. We will drop that level
```

```python
categorical = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
```

```python
categorical
```

```python
for feature in categorical:
    x = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=x[x>0.01].index # to get the index name
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for feature in categorical:
    a = le.fit_transform(dataset[feature])
    dataset[feature] = a
```
